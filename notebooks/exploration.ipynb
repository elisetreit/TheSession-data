```python
a = df.merge(names, how = 'inner', on = 'tune_id').groupby("tune_id").agg(alias_num = ('alias', 'nunique'), setting_num = ('setting_id', 'nunique'))
```

```python
a["name_setting"] = a['alias_num'] / a['setting_num']
```

```python
a["setting_name"] = a['setting_num']  / a['alias_num']
```
